```python
import pandas as pd
datos = {
    "Nombre": ["Ana", "Luis", "Carlos", None, "Marta"],
    "Edad": [25, None, 40, 35, -5],
    "Salario": [50000, 55000, 60000, 58000, None]
}
df = pd.DataFrame(datos)
# Ver valores faltantes
print(df.isnull().sum())
# Ver datos fuera de rango
print(df[df["Edad"] < 0])
```

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

# Normalización de la columna "Salario"
scaler = MinMaxScaler()
df["Salario_Normalizado"] = scaler.fit_transform(df[["Salario"]])

# Codificación de la columna "Nombre"
encoder = LabelEncoder()
df["Nombre_Codificado"] = encoder.fit_transform(df["Nombre"].astype(str))

# Mostrar resultado
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Histograma de edades
sns.histplot(df["Edad"].dropna(), bins=5, kde=True)
plt.show()
# Boxplot de salarios
sns.boxplot(x=df["Salario"])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Crear DataFrame
datos = {
    "Cliente": [1, 2, 3, 4, 5],
    "Edad": [25, 40, 32, 22, 35],
    "Plan": ["Básico", "Premium", "Estándar", "Básico", "Premium"],
    "Consumo_GB": [5, 50, 10, 7, 45],
    "Reclamos": [1, 0, 2, 3, 0],
    "Estado_Cuenta": ["Pagado", "Pagado", "Moroso", "Moroso", "Pagado"]
}
df = pd.DataFrame(datos)
# Inspeccionar los datos
print(df.info())
print(df.describe())
# Ver distribución de consumo de datos
sns.histplot(df["Consumo_GB"], bins=5, kde=True)
plt.show()
# Gráfico de barras de Estado de Cuenta
sns.countplot(x=df["Estado_Cuenta"])
plt.show()
```
