```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear DataFrame con datos de ejemplo
df = pd.DataFrame({
    "Edad": [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos": [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000]
})

# Medidas estadísticas básicas
print(df["Edad"].describe())

# Histograma para analizar la distribución
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.show()

# Boxplot para detectar outliers
sns.boxplot(x=df["Edad"])
plt.title("Boxplot de Edades")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear DataFrame
df = pd.DataFrame({
    "Edad": [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos": [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000]
})

# Matriz de correlación
print(df.corr())

# Mapa de calor de correlación
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación")
plt.show()

# Gráfico de dispersión
sns.scatterplot(x=df["Edad"], y=df["Ingresos"])
plt.title("Relación entre Edad e Ingresos")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Crear DataFrame con datos ficticios
df = pd.DataFrame({
    "Edad": [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos": [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000],
    "Experiencia": [2, 10, 7, 15, 3, 12, 4, 9, 11, 20]
})

# Normalizar los datos
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Aplicar PCA para reducir a 2 componentes
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(df_pca, columns=["PC1", "PC2"])

# Visualizar los resultados
sns.scatterplot(x=df_pca["PC1"], y=df_pca["PC2"])
plt.title("Proyección de Datos en Componentes Principales")
plt.show()
```

```python
from sklearn.cluster import KMeans

# Aplicar K-Means con 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(df_scaled)

# Visualizar clusters
sns.scatterplot(x=df["Ingresos"], y=df["Experiencia"], hue=df["Cluster"], palette="viridis")
plt.title("Clusters según Ingresos y Experiencia")
plt.show()
```
