```python
import pandas as pd

# Crear DataFrame

datos = {
"Cliente": [1, 2, 3, 4, 5],
"Edad": [25, -40, 32, 22, 35],
"Salario": [50000, 60000, 70000, None, 80000],
"Estado_Civil": ["Soltero", "Casado", "Divorciado", "Soltero", "Viudo"],

"Credito_Aprobado": ["Sí", "No", "Sí", "No", "Sí"]
}
df = pd.DataFrame(datos)
# Ver problemas en los datos

print(df.isnull().sum()) # Valores faltantes

print(df[df["Edad"] < 0]) # Valores atípicos
```

```python
# Crear un DataFrame con distintos tipos de datos
df = pd.DataFrame({
    "Edad": [25, 30, 35, 40, 45],
    "Ingreso": [50000, 60000, 55000, 75000, 90000],
    "Estado_Civil": ["Soltero", "Casado", "Viudo", "Casado", "Soltero"],
    "Nivel_Educativo": ["Secundaria", "Universitaria", "Secundaria", "Primaria", "Universitaria"]
})

# Inspeccionar los tipos de datos
print(df.dtypes)

from sklearn.preprocessing import OrdinalEncoder

# Definir orden de categorías
orden_educativo = [["Primaria", "Secundaria", "Universitaria"]]

# Aplicar codificación ordinal
encoder = OrdinalEncoder(categories=orden_educativo)
df["Nivel_Educativo_Cod"] = encoder.fit_transform(df[["Nivel_Educativo"]])

print(df[["Nivel_Educativo", "Nivel_Educativo_Cod"]])

```

```python
from sklearn.preprocessing import OneHotEncoder

# Aplicar One-Hot Encoding
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_data = encoder.fit_transform(df[["Estado_Civil"]])

# Convertir a DataFrame
df_encoded = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(["Estado_Civil"]))
df = pd.concat([df, df_encoded], axis=1)

print(df.head())
```

```python
import pandas as pd

from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

# Crear DataFrame

data = {
"ID": [1, 2, 3, 4],
"Nombre": ["Juan", "María", "Pedro", "Ana"],
"Edad": [32, 28, 40, 35],
"Categoria_Cliente": ["Oro", "Plata", "Platino", "Oro"],

"Estado_Civil": ["Soltero", "Casado", "Viudo", "Casado"]
}

df = pd.DataFrame(data)

# Codificación ordinal para "Categoría Cliente"

orden_categoria = [["Plata", "Oro", "Platino"]]

encoder_ordinal = OrdinalEncoder(categories=orden_categoria)
df["Categoria_Cod"] = encoder_ordinal.fit_transform(df[["Categoria_Cliente"]])

# One-Hot Encoding para "Estado Civil"

encoder_onehot = OneHotEncoder(sparse_output=False, drop="first")

encoded_estado = encoder_onehot.fit_transform(df[["Estado_Civil"]])

df_encoded = pd.DataFrame(
    encoded_estado,
    columns=encoder_onehot.get_feature_names_out(["Estado_Civil"])
)

df = pd.concat([df, df_encoded], axis=1)

print(df.head())
```

```python
import pandas as pd

from sklearn.preprocessing import OrdinalEncoder

# Crear DataFrame
df = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Nombre": ["Juan", "María", "Pedro", "Ana"],
    "Edad": [32, 28, 40, 35],
    "Categoria_Cliente": ["Oro", "Plata", "Platino", "Oro"],
    "Ingresos": [45000, 55000, 70000, 60000]
})

# Estadísticas descriptivas
print(df.describe())
print(df.describe(include='all'))

# Codificación ordinal de "Categoría Cliente"

orden_categoria = [["Plata", "Oro", "Platino"]]

encoder = OrdinalEncoder(categories=orden_categoria)
df["Categoria_Cod"] = encoder.fit_transform(df[["Categoria_Cliente"]])

# Calcular correlación

print(df[["Edad", "Ingresos"]].corr())

print(df)
```
