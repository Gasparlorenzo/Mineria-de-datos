```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear DataFrame
df = pd.DataFrame({
    "Cliente":   [1, 2, 3, 4, 5, 6, 7, 8],
    "Edad":      [25, 40, 35, 50, 30, 45, 38, 29],
    "Ingresos":  [50000, 60000, 70000, 80000, 45000, 85000, 72000, 48000],
    "Compras":   [5, 15, 10, 20, 7, 18, 12, 6],
    "Categoria": ["A", "B", "C", "A", "B", "C", "A", "B"]
})
```

```python
df
```

## Análisis Exploratorio de Datos (EDA)
- Generar histogramas para Edad e Ingresos.
- Crear un boxplot para visualizar valores atípicos en Compras.
- Generar un scatter plot para analizar la relación entre Ingresos y Compras.

# **Clasificacion de datos segun su tipo**
Variables Cuantitativas (numéricas)

Continuas
- Ingresos → puede tomar cualquier valor numérico dentro de un rango (50000, 60000, etc.)


Discretas

- Cliente → número entero que identifica a cada cliente
- Edad → años completos (25, 40, 35...)
- Compras → cantidad entera de compras realizadas (5, 15, 10...)

Variables Cualitativas (categóricas)

Nominales (no tienen orden)

- Categoria → etiquetas A, B, C

```python
plt.figure(figsize=(7,6))
# Histograma de Edad
plt.figure(figsize=(6,4))
sns.histplot(df["Edad"], bins=5, kde=True, color="#378ADD", edgecolor="white")

plt.title("Distribución de Edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")

plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()


# Histograma de Ingresos
plt.figure(figsize=(6,4))
sns.histplot(df["Ingresos"], bins=5, kde=True, color="#E67E22", edgecolor="white")

plt.title("Distribución de Ingresos")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")

plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Compras"], color="#1D9E75", linecolor="black", linewidth=0.8)
plt.title("Distribución de compras por cliente")
plt.xlabel("Cantidad de compras")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.scatterplot(
    x=df["Ingresos"],
    y=df["Compras"],
    hue=df["Categoria"],      # color por categoría
    palette="Set2",           # colores suaves y claros
    s=100,                    # tamaño de puntos
    edgecolor="black",        # borde para resaltar
    alpha=0.8
)

plt.title("Relación entre Ingresos y Compras", fontsize=12)
plt.xlabel("Ingresos")
plt.ylabel("Compras")
plt.grid(linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
correlacion = df["Ingresos"].corr(df["Compras"])
print(f"Correlación: {correlacion:.2f}")
```

📊 Conclusiones del Análisis Exploratorio

A partir del análisis realizado sobre el dataset, se pueden obtener las siguientes conclusiones:

En primer lugar, al analizar la variable Edad, se observa que los clientes se encuentran en un rango relativamente amplio, aunque con mayor concentración en edades intermedias. Esto indica que no hay un grupo etario dominante muy marcado, sino una distribución bastante equilibrada.

En cuanto a los Ingresos, se identifica una dispersión considerable, lo que sugiere la presencia de clientes con distintos niveles económicos. Esta variabilidad resulta clave para analizar el comportamiento de compra.

El boxplot de Compras muestra que la mayoría de los valores se concentran en niveles bajos y medios, aunque existen algunos valores más altos que podrían considerarse extremos dentro del conjunto. Sin embargo, no se observan outliers demasiado alejados, lo que indica una distribución relativamente estable.

Por otro lado, el gráfico de dispersión entre Ingresos y Compras evidencia una clara relación positiva. Esto significa que, en general, a medida que aumentan los ingresos, también lo hace la cantidad de compras. Este patrón sugiere que el poder adquisitivo influye directamente en el comportamiento de consumo.

Además, al incorporar la variable Categoría, se puede observar que los distintos grupos de clientes se distribuyen dentro de esta tendencia general, aunque sin una separación completamente definida. Esto indica que, si bien la categoría aporta información adicional, no es el único factor que explica el comportamiento de compra.

## Visualización Avanzada
- Crear un Pairplot con las variables numéricas.
- Generar un Heatmap de correlaciones entre todas las variables numéricas.
- Utilizar un Violin Plot para comparar la distribución de Compras según la Categoría.

```python
sns.pairplot(df[["Edad", "Ingresos", "Compras"]], corner=True)
plt.show()
```

```python
# Mapa de calor de correlaciones
correlaciones = df.corr(numeric_only=True)
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación de Variables")
plt.show()
```

```python
# Violin Plot de Compras por Categoría
sns.violinplot(x="Categoria", y="Compras", data=df, palette="muted")
plt.title("Distribución de Compras por Categoría")
plt.show()
```

## Reflexión sobre las visualizaciones avanzadas

El uso de visualizaciones avanzadas, como el pairplot, el heatmap y el violin plot, permitió profundizar el análisis de los datos más allá de las observaciones iniciales.

En el caso del pairplot, fue posible visualizar de manera conjunta las relaciones entre las variables numéricas, lo que facilitó identificar patrones y confirmar la relación positiva entre ingresos y compras. Además, permitió observar la dispersión de los datos desde diferentes perspectivas.

Por su parte, el heatmap de correlaciones brindó una visión más precisa y cuantitativa de la relación entre variables, complementando lo observado en el gráfico de dispersión. Esto permitió validar que existe una fuerte correlación entre ingresos y compras, reforzando una de las principales conclusiones del análisis.

Finalmente, el violin plot aportó una forma más completa de analizar la distribución de compras según la categoría, ya que combina información de densidad y dispersión. Esto permitió comparar los distintos grupos de clientes y entender mejor cómo se distribuyen los valores dentro de cada categoría.
