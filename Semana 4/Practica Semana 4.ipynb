```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear datos ficticios
df = pd.DataFrame({
    "Edad": [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos": [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000],
    "Categoría": ["Joven", "Adulto", "Adulto", "Adulto", "Senior", "Senior", "Joven", "Adulto", "Adulto", "Senior"]
})

# Histograma de edades
sns.histplot(df["Edad"], bins=5, kde=True)
plt.title("Distribución de Edades")
plt.show()

# Boxplot de ingresos
sns.boxplot(x=df["Ingresos"])
plt.title("Boxplot de Ingresos")
plt.show()

# Scatter Plot entre Edad e Ingresos
sns.scatterplot(x=df["Edad"], y=df["Ingresos"], hue=df["Categoría"])
plt.title("Relación entre Edad e Ingresos")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear DataFrame
df = pd.DataFrame({
    "Categoría": ["Junior", "Semi-Senior", "Senior"],
    "Salario_Promedio": [35000, 60000, 90000]
})

# Gráfico de barras con Matplotlib
plt.bar(df["Categoría"], df["Salario_Promedio"], color=["#1f77b4", "#ff7f0e", "#2ca02c"])
plt.xlabel("Categoría de Empleados")
plt.ylabel("Salario Promedio")
plt.title("Salario Promedio por Categoría")
plt.show()

# Crear DataFrame con edades e ingresos
df2 = pd.DataFrame({
    "Edad": [25, 32, 40, 35, 50, 45, 29, 37, 41, 60],
    "Ingresos": [30000, 45000, 60000, 55000, 80000, 75000, 40000, 58000, 72000, 90000]
})

# Scatterplot con Seaborn
sns.scatterplot(x=df2["Edad"], y=df2["Ingresos"],color="#6C5CE7",s=100)
plt.title("Relación entre Edad e Ingresos",fontsize=14)
plt.grid(color="lightgray", linestyle="--", linewidth=0.7, alpha=0.7)
plt.show()
```

Objetivo: Aplicar gráficos avanzados para extraer insights de un dataset.

Instrucciones:

- Carga el dataset de diamantes de Seaborn (diamonds).
- Crea un Pairplot para analizar la relación entre precio, quilates y profundidad.
- Genera un Heatmap para visualizar la correlación entre las variables numéricas.
- Crea un Violin Plot para comparar la distribución del precio en cada categoría de corte (cut).
-Escribe una conclusión basada en los gráficos.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Cargar el dataset de diamantes de Seaborn
df = sns.load_dataset("diamonds")
print(df.head())
```

```python
# Resumen Estadístico
df.describe()
```

```python
print(df.info())
```

Content
price price in US dollars (\$326--\$18,823)

carat weight of the diamond (0.2--5.01)

cut quality of the cut (Fair, Good, Very Good, Premium, Ideal)

color diamond colour, from J (worst) to D (best)

clarity a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

x length in mm (0--10.74)

y width in mm (0--58.9)

z depth in mm (0--31.8)

depth total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)

table width of top of diamond relative to widest point (43--95)

```python

# 1. Pairplot entre precio, quilates y profundidad
sns.pairplot(df[["price", "carat", "depth"]])
plt.show()

# 2. Heatmap de correlaciones
correlaciones = df.corr(numeric_only=True)
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación de Diamantes")
plt.show()

# 3. Violin Plot para comparar precio según corte
sns.violinplot(x="cut", y="price", data=df, palette="muted")
plt.title("Distribución del Precio por Tipo de Corte")
plt.show()
```

1. ¿Qué variables están más correlacionadas con el precio?
Las variables carat (quilates), x (largo), y (ancho) y z (profundidad en mm) son las que tienen mayor correlación con price. La más fuerte es carat, con una correlación cercana a 0.92. Esto tiene sentido: a mayor peso del diamante, mayor precio.

2. ¿Cómo varía la distribución del precio en los diferentes tipos de corte?
Sorprendentemente, los diamantes con corte Fair (el peor corte) tienden a tener precios medianos más altos que los de corte Ideal (el mejor). Esto se explica porque los diamantes de corte Fair suelen ser más grandes en quilates, lo que compensa la menor calidad del corte. El Violin Plot muestra que todos los tipos de corte tienen una distribución sesgada hacia precios bajos, con una cola larga hacia precios altos.

3. ¿Qué otros insights podés extraer de estos gráficos?

El Pairplot muestra que carat y price tienen una relación positiva no lineal (curva, no recta), lo que sugiere que el precio crece más rápido a medida que aumentan los quilates.
El Heatmap revela que x, y y z están altamente correlacionadas entre sí (cercano a 1.0), lo que indica multicolinealidad — no aportan información independiente.
La variable depth tiene correlación muy baja con el precio (~0.01), por lo que no es un buen predictor.
